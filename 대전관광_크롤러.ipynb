```python
# 네이버에서 검색어 입력받아 검색 한 후 블로그 메뉴를 선택하고
# 오른쪽에 있는 검색옵션 버튼을 눌러서
# 정렬 방식과 기간을 입력하기

#Step 0. 필요한 모듈과 라이브러리를 로딩합니다.
import sys # 시스템
import os  # 시스템

import pandas as pd  # 판다스 : 데이터분석 라이브러리
import numpy as np   # 넘파이 : 숫자, 행렬 데이터 라이브러리

from bs4 import BeautifulSoup    # html 데이터를 전처리
from selenium import webdriver   # 웹 브라우저 자동화
import time    # 서버와 통신할 때 중간중간 시간 지연. 보통은 1초
from tqdm import tqdm_notebook   # for문 돌릴 때 진행상황을 %게이지로 알려준다.
```

```python
#Step 1. 크롬 웹브라우저 실행
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")

driver = webdriver.Chrome(chrome_options=options)

driver.get('https://www.daejeon.go.kr/tou/index.do')
time.sleep(2)  # 2초간 정지
```

```python
#Step 2.

driver.find_element_by_link_text("관광명소").click( )
time.sleep(1)

driver.find_element_by_link_text("분야별관광명소").click( )
time.sleep(1)
```

```python
list = driver.find_elements_by_xpath('//*[@id="txt"]/div[1]/ul//li')
list
```

```python
page_count = driver.find_element_by_xpath('//*[@id="txt"]/div[1]/div[2]/div[1]/div[1]/a[last ()]').text
page_count
```

```python
dict = {}
dic_count=0

for i in range(2, int(page_count)+1):
    list = driver.find_elements_by_xpath('//*[@id="txt"]/div[1]/ul//li')
    for element in list:
        try:
            # 사진
            img = element.find_element_by_xpath('div[1]/a/img').get_attribute('src')
            print(img)

            # 장소 이름
            name = element.find_element_by_xpath('a/h4').text[4:]
            print(name)

            # 주소
            addr = element.find_element_by_xpath('ul/li').text[6:]
            print(addr)

            target_info = {}
            target_info['사진'] = img
            target_info['장소 이름'] = name
            target_info['주소'] = addr

            dict[dic_count] = target_info
            dic_count = dic_count+1

            time.sleep(1)
        except:
            pass
        
    result = 'fn_link_page(' + str(i) + ')'
    driver.execute_script(result)

    time.sleep(1)
```

```python
# 판다스로 만들기
import pandas as pd
result_df = pd.DataFrame.from_dict(dict, 'index')
result_df
```

```python
# 엑셀로 저장하기
result_df.to_excel("대전관광명소.xlsx", encoding='utf-8-sig')
```
