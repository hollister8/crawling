```python
import pandas as pd
import numpy as np

from selenium import webdriver  # 라이브러리(모듈) 가져오라
from selenium.webdriver import ActionChains as AC # 웹브라우저 동작
from tqdm import tqdm
from tqdm.notebook import tqdm
import re
from time import sleep
import time
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
```

```python
# 데이터 수집할 키워드 지정
keyword = "송촌동"
keyword
```

```python
# 크롬창 띄우기
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")

driver = webdriver.Chrome(chrome_options=options)

#driver = webdriver.Chrome("chromedriver.exe")  # 맥 : /Users/aiden/Desktop/chromedriver
driver.get("https://m.place.naver.com/my")
time.sleep(3)
# 로그인은 수동
```

```python
last_page_height = driver.execute_script("return document.documentElement.scrollHeight")
count = 0
while True:
    driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
    time.sleep(2.0)
    new_page_height = driver.execute_script("return document.documentElement.scrollHeight")
    count = count + 1
    if new_page_height == last_page_height:
        break
    if count >= 3:
        break
    last_page_height = new_page_height
```

```python
list = driver.find_elements_by_xpath('//*[@id="app-root"]/div[3]/div/ul//li')
```

```python
urlvalue = []
for element in list:
    
    try:
        a = element.find_element_by_xpath('div[4]/div[1]/div[2]/div[1]').get_attribute('data-cid')
        urlvalue.append(a)
    except:
        pass
        
urlvalue
```

```python
dict = {}
dic_count = 0

for i in urlvalue:

    #driver = webdriver.Chrome("chromedriver.exe")  # 맥 : /Users/aiden/Desktop/chromedriver
    driver.get('https://m.place.naver.com/place/' + i)
    
    # 가게 이름
    name = driver.find_element_by_xpath('//*[@id="_title"]/span[1]').text
    print(name)

    # 별점
    try:
        star = driver.find_element_by_xpath('//*[@class="_1Y6hi _1A8_M"]/em[1]').text
        print(star)
    except:
        pass
    
    # '리뷰' 클릭
    driver.find_element_by_link_text("리뷰").click()
    time.sleep(2)
    
    last_page_height = driver.execute_script("return document.documentElement.scrollHeight")
    count = 0
    while True:
        driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
        time.sleep(2.0)
        new_page_height = driver.execute_script("return document.documentElement.scrollHeight")
        count = count + 1
        if new_page_height == last_page_height:
            break
        if count >= 2:
            break
        last_page_height = new_page_height
        
    # 리뷰수
    review_count = driver.find_element_by_css_selector('.place_section_count')
    review_count = review_count.text
    print(review_count)
    
    # 리뷰 내용
    review_list = []
    reviews = driver.find_elements_by_xpath('//*[@class="PVBo8"]/a//span')
    for review in reviews:
        print(review.text)
        review_list.append(review.text)

    try:
        driver.find_element_by_link_text("더보기").click()
        time.sleep(2)
        
    except:
        pass
    
    target_info = {}
    target_info['가게 이름'] = name
    target_info['별점'] = star
    target_info['리뷰수'] = review_count
    target_info['리뷰 내용'] = review_list
    
    dict[dic_count] = target_info
    dic_count = dic_count+1
```

```python
# 판다스로 만들기
import pandas as pd
result_df = pd.DataFrame.from_dict(dict, 'index')
result_df
# result_df.rename(columns={result_df.columns[0] : name}, inplace=True)
```

```python
# 파일로 저장
result_df.to_excel("naver_place_({}).xlsx".format(keyword))
```
